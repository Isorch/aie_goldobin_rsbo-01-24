```python
# 1. Импортирование библиотеки Numpy под общепринятым сокращением np
import numpy as np
```

```python
# 2. Примеры создания массивов
arr_1 = np.array([1, 2, 3, 4]) # демонстрация преобразования обычного списка в массив numpy
print("arr_1: ", arr_1, arr_1.dtype, '\n')

arr_2 = np.array([[234, 234.54, "hello"], [34, 532, 123]]) # преобразование двумерного списка
print("arr_2: ", arr_2, arr_2.dtype, '\n')

zero = np.zeros((3, 3)) # создание массива numpy из нулей 3x3. Можно задать одномерный массив без передачи кортежа в функцию, например np.zeros(6)
print("zero: ", zero, '\n')

one = np.ones((2, 2, 3)) # создание массива numpy из единиц
print("one: ", one, '\n')

arng_1 = np.arange(10) # генерация массива из последовательности
print("arng_1: ", arng_1, '\n')

arng_2 = np.arange(0, 3.4, 0.3) # в отличие от обычного range работает с дробными чмслами
print("arng_2: ", arng_2, '\n')

array_full = np.full((3, 3), 10) # функция создания массива с заданным размером и наполняемым числом
print("array_full: ", array_full, '\n')

arr_empt = np.empty((2, 2)) # создание пустого массива (без инициализации значений)
print("arr_empt: ", arr_empt, '\n')
```

```python
# 3. Демонстрация атрибутов
print("Форма массива one: ", one.shape)
print("Тип данных массива one: ", one.dtype)
print("Кол-во измерений массива one: ", one.ndim)
print("Общее кол-во элементов массива one: ", one.size)
```

```python
# 4. Примеры индексирования и срезов
print(arr_2[1, 1]) # Получение отдельного элемента из массива arr_2 (индексация в квадратных скобках через запятую)
print(arr_2[:2, 1:3]) # Извлечение подмассива с помощью среза
```

```python
# 5. Примеры векторных оперцаий и базовых статистик
first = np.arange(1, 9)
second = np.arange(3, 11)

sum = first + second # Поэлементарное сложение
print(sum)

diff = second - first # Поэлементарное вычитание
print(diff)

mult = first * second # Поэлементарное умножение
print(mult)

print(np.mean(second)) # среднее массива
print(np.median(second)) # медиана массива
print(np.std(second)) # стандартное отклонение массива
```
